## Practica Cargo Datos

```python
#%pip install ipykernel
#%pip install numpy
#%pip install pandas

import numpy as np


url = 'https://gist.githubusercontent.com/ahcamachod/41b8a65c5e5b58125401deafb68af460/raw/8f328f1777a4d628015f827240457a1f8078ab02/manzanas.csv'

np.loadtxt(url, delimiter=',', usecols = np.arange(1,88,1))
datos = np.loadtxt(url, delimiter=',', usecols = np.arange(1,88,1))

# Metodos basicos para explorar Arrays
# 1. shape: devuelve la forma del array
print(datos.shape)
# 2. ndim: devuelve el numero de dimensiones del array
print(datos.ndim)
# 3. size: devuelve el numero total de elementos del array
print(datos.size)
# 4. dtype: devuelve el tipo de datos del array
print(datos.dtype)

datos.T
```

## Practica Visualización y selección de datos

```python
datos_transpuestos = datos.T
fechas = datos_transpuestos[:,0]
print(fechas)
print("------------------------------")

precios = datos_transpuestos[:,1:]
print(precios)
print("------------------------------")

import matplotlib.pyplot as plt

##Graficando por lugar
fechas = np.arange(1, 88, 1)
Moscu = precios[:,0]
Kaliningrado = precios[:,1]
Petersburgo = precios[:,2]
Krasnodar = precios[:,3]
Ekaterimburgo = precios[:,4]

plt.plot(fechas, Moscu, label='Moscu')
plt.plot(fechas, Kaliningrado, label='Kaliningrado')
plt.plot(fechas, Petersburgo, label='Petersburgo')
plt.plot(fechas, Krasnodar, label='Krasnodar')
plt.plot(fechas, Ekaterimburgo, label='Ekaterimburgo')
plt.title('Precios de manzanas en Rusia')
```

## Comparación entre arrays

```python
moscu_1 = Moscu[0:12]
moscu_2 = Moscu[12:24]
moscu_3 = Moscu[24:36]
moscu_4 = Moscu[36:48]

plt.plot(np.arange(1, 13, 1), moscu_1, label='2013')
plt.plot(np.arange(1, 13, 1), moscu_2, label='2014')
plt.plot(np.arange(1, 13, 1), moscu_3, label='2015')
plt.plot(np.arange(1, 13, 1), moscu_4, label='2016')

plt.title('Precios de manzanas en Moscu')
plt.xlabel('Meses')
plt.ylabel('Precio')
plt.legend()
plt.show()
```

## Tratamiento de Nan

```python
Kaliningrado

np.isnan(Kaliningrado)

#Sacando el Nan hay que sumar el anterior y el posterior y dividirlo entre 2 y asignarlo a la posicion del NaN
for i in range(len(Kaliningrado)):
    if np.isnan(Kaliningrado[i]):
        Kaliningrado[i] = (Kaliningrado[i-1] + Kaliningrado[i+1]) / 2
print(Kaliningrado)

# Precio promedio manzana por año en Kaliningrado
kaliningrado_1 = Kaliningrado[0:12]
kaliningrado_2 = Kaliningrado[12:24]
kaliningrado_3 = Kaliningrado[24:36]
kaliningrado_4 = Kaliningrado[36:48]

plt.plot(np.arange(1, 13, 1), kaliningrado_1, label='2013')
plt.plot(np.arange(1, 13, 1), kaliningrado_2, label='2014')
plt.plot(np.arange(1, 13, 1), kaliningrado_3, label='2015')
plt.plot(np.arange(1, 13, 1), kaliningrado_4, label='2016')
plt.title('Precios de manzanas en Kaliningrado')
plt.xlabel('Meses')
plt.ylabel('Precio')
plt.legend()
plt.show()


```

## Diferencia en Arrays

```python
plt.plot(fechas, Moscu, label='Moscu')
x = fechas
y = 0.5 * x + 78
plt.plot(x, y, label='y=0.5x+78')
np.sum(np.power(Moscu -y,2))
plt.plot(fechas, Moscu, label='Moscu')
```

## Regresión lineal

```python
n = np.size(Moscu)
y = Moscu
x = fechas
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
b = (np.sum(y) - a*np.sum(x))/n
print(f'La pendiente es {a} y la interseccion es {b}')

y = a * x + b

np.linalg.norm(Moscu - y)
```

## Dibujando circulo

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una secuencia de valores de x de -1 a 1
x = np.arange(-1, 1, 0.0001)

# Implementación de la fórmula
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

# Graficar el gráfico con las dos partes del círculo
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r' )

# Agregar el título del gráfico y las etiquetas de los ejes x e y
plt.title("Círculo")
plt.xlabel("Eje x")
plt.ylabel("Eje y")

# Mostrar el gráfico
plt.show()
```

## Trabajando numeros aleatorios (float / enteros)

```python
np.random.randint(1, 10, 10)
np.random.uniform(1, 10, 10)
np.random.normal(0, 1, 10)

## 3 Ejemplo con pendientes

pendientes = np.random.uniform(0, 1, 10)
intersecciones = np.random.uniform(0, 1, 10)
x = np.arange(0, 1, 0.01)
y = np.zeros((10, 100))
for i in range(10):
    y[i] = pendientes[i] * x + intersecciones[i]
    plt.plot(x, y[i], label=f'pendiente {pendientes[i]} interseccion {intersecciones[i]}')
plt.title('Pendientes')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# EL mejor valor de la pendiente es el que tiene el menor error cuadratico medio

for i in range(10):
    y[i] = pendientes[i] * x + intersecciones[i]
    plt.plot(x, y[i], label=f'pendiente {pendientes[i]} interseccion {intersecciones[i]}')
plt.title('Pendientes')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
